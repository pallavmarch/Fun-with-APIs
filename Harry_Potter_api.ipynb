```python
import requests
import pandas as pd
from tqdm import tqdm

url='https://hp-api.onrender.com/api/characters'

response=requests.get(url)
data=response.json()

characters=[]
```

```python
for char in tqdm(data,colour='blue'):
    name=char['name']
    alternate_names=char['alternate_names']
    actor=char['actor']
    species=char['species']
    gender=char['gender']
    house=char['house']
    dateOfBirth=char['dateOfBirth']
    yearOfBirth=char['yearOfBirth']
    wizard=char['wizard']
    ancestry=char['ancestry']
    eyeColour=char['eyeColour']
    hairColour=char['hairColour']
    patronus=char['patronus']
    hogwartsStudent=char['hogwartsStudent']
    hogwartsStaff=char['hogwartsStaff']
    alive=char['alive']
    wand = char.get('wand', {})
    wand_wood = wand.get('wood')
    wand_core = wand.get('core')
    wand_length = wand.get('length')

    characters.append({
        'name':name,
        'alternate_names':alternate_names,
        'actor':actor,
        'species':species,
        'gender':gender,
        'house':house,
        'dateOfBirth':dateOfBirth,
        'yearOfBirth':yearOfBirth,
        'wizard':wizard,
        'ancestry':ancestry,
        'eyeColour':eyeColour,
        'hairColour':hairColour,
        'wand_wood': wand_wood,
        'wand_core': wand_core,
        'wand_length': wand_length,
        'patronus':patronus,
        'hogwartsStudent':hogwartsStudent,
        'hogwartsStaff':hogwartsStaff,
        'alive':alive
         })

df=pd.DataFrame(characters,columns=['name','alternate_names','actor','species','gender','house','dateOfBirth','yearOfBirth','wizard','ancestry','eyeColour','hairColour','wand_wood','wand_core','wand_length','patronus','hogwartsStudent','hogwartsStaff','actor','alive'])
df.to_csv('Harry_Potter_characters.csv',index=False)

df.head()
```
