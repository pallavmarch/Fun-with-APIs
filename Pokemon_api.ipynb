```python
import requests
from tabulate import tabulate


name = 'pikachu'
url='https://pokeapi.co/api/v2/pokemon/'+name

response=requests.get(url)
data=response.json()
```

```python
data.keys()
#data.values()
```

```python
name=data['name']
height=data['height']
weight=data['weight']
base_experience=data['base_experience']


abilities=[]
for ability in data['abilities']:
    abilities.append(ability['ability']['name'])

types=[]
for type in data['types']:
    types.append(type['type']['name'])

stats={}
for stat in data['stats']:
    stats[stat['stat']['name']]=stat['base_stat']

hp=stats.pop('hp')
attack=stats.pop('attack')
defense=stats.pop('defense')
special_attack=stats.pop('special-attack')
special_defense=stats.pop('special-defense')
speed=stats.pop('speed')
```

```python
table = [
    ["Name", data['name'].title()],
    ["Height", data['height']],
    ["Weight", data['weight']],
    ["Base Experience", data['base_experience']],
    ["Abilities", abilities],
    ["Types", types],
    ["HP", hp],
    ["Attack", attack],
    ["Defense", defense],
    ["Special Attack", special_attack],
    ["Special Defense", special_defense],
    ["Speed", speed]
]

print(f"\nPokémon Information for '{name.title()}':\n")
print(tabulate(table, headers=["Attribute", "Value"], tablefmt="fancy_grid"))
```

# Fetching Data

```python
import pandas as pd
from tqdm import tqdm


url = " https://pokeapi.co/api/v2/pokemon?limit=100000&offset=0"  # covers all Pokémon
response=requests.get(url)
data=response.json()['results']

all_pokemon=[]
for pokemon in tqdm(data,colour='blue'):
    url=pokemon['url']
    response=requests.get(url)
    data=response.json()

    name=data['name']
    height=data['height']
    weight=data['weight']
    base_experience=data['base_experience']


    abilities=[]
    for ability in data['abilities']:
        abilities.append(ability['ability']['name'])

    types=[]
    for type in data['types']:
        types.append(type['type']['name'])

    stats={}
    for stat in data['stats']:
        stats[stat['stat']['name']]=stat['base_stat']

    hp=stats.pop('hp')
    attack=stats.pop('attack')
    defense=stats.pop('defense')
    special_attack=stats.pop('special-attack')
    special_defense=stats.pop('special-defense')
    speed=stats.pop('speed')


    all_pokemon.append({
        "Name": name.title(),
        "Height": height,
        "Weight": weight,
        "Base Experience": base_experience,
        "Abilities": abilities,
        "Types": types,
        "HP": hp,
        "Attack": attack,
        "Defense": defense,
        "Special Attack": special_attack,
        "Special Defense": special_defense,
        "Speed": speed
    })


df=pd.DataFrame(all_pokemon)
df.to_csv('pokemon.csv',index=False)
df.head()
```
