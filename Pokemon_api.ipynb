```python
import requests
name = 'pikachu'
url='https://pokeapi.co/api/v2/pokemon/'+name

data=requests.get(url)
data=data.json()
```

```python
abilities = ', '.join([a['ability']['name'] for a in data['abilities']])
types = ', '.join([t['type']['name'] for t in data['types']])

table = [
    ["Name", data['name'].title()],
    ["Height", data['height']],
    ["Weight", data['weight']],
    ["Base Experience", data['base_experience']],
    ["Abilities", abilities],
    ["Types", types]
]

print(f"\nPokémon Information for '{name.title()}':\n")
print(tabulate(table, headers=["Attribute", "Value"], tablefmt="fancy_grid"))

```

```python
import pandas as pd
from tqdm import tqdm


base_url = "https://pokeapi.co/api/v2/pokemon?limit=2000"  # covers all Pokémon
response = requests.get(base_url)
pokemon_list = response.json()['results']

# Step 2: Fetch details for each Pokémon
all_data = []

for pokemon in tqdm(pokemon_list, desc="Fetching Pokémon data"):
    url = pokemon['url']
    res = requests.get(url)

    if res.status_code == 200:
        data = res.json()
        name = data['name'].title()
        height = data['height']
        weight = data['weight']
        base_exp = data.get('base_experience', None)
        abilities = ', '.join([a['ability']['name'] for a in data['abilities']])
        types = ', '.join([t['type']['name'] for t in data['types']])

        all_data.append({
            "Name": name,
            "Height": height,
            "Weight": weight,
            "Base Experience": base_exp,
            "Abilities": abilities,
            "Types": types
        })

# Step 3: Save as CSV or display
df = pd.DataFrame(all_data)
df.to_csv("pokemon_dataset.csv", index=False)
print("Saved full Pokémon dataset to 'pokemon_dataset.csv'")
```
